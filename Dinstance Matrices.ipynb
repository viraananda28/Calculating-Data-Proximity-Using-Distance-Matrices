```python
import numpy as np
import pandas as pd
#df = pd.read_csv("batterytrain.csv", nrows=100)
df = pd.read_csv("batterytrain.csv")
df = df.iloc[0:100]
df
```

## No 1. Memperoleh mix atribut dari jarak matriks

Mengidentifikasi kolom yang numerik, ordinal, atau data biner.

```python
# Definisi jarak matrix
distance = np.zeros([len(df),len(df)])

# Definisi kategori kolom (numerik, ordinal, biner)
num_columns = ['battery_power', 'fc', 'int_memory', 'm_dep', 'mobile_wt', 'px_height', 'px_width', 'ram', 'sc_h', 'sc_w', 
              'talk_time', 'pc']
ord_columns = ['clock_speed', 'n_cores', 'price_range']
bin_columns = ['blue', 'dual_sim', 'four_g', 'three_g', 'touch_screen', 'wifi']

# Iterasi jarak kolom
for i in range(0, 100):
    
    # Iterasi melalui nama kolom dan penjumlahan masing-masing kolom
    for col in df.columns:
        
        # Hitung dengan jarak numerik
        if col in num_columns:
            distance[:,i] += abs(df[col][i]-df[col])/(np.max(df[col])-np.min(df[col]))
            
        # Hitung dengan jarak ordinal
        elif col in ord_columns:
            distance[:,i] += df[col]!=df[col][i]
        # Hitung dengan jarak biner
        else:
            distance[:,i] += df[col]!=df[col][i]
            
    # Hasil penjumlahan dibagi dg jumlah dari kolom
    distance[:,i] = distance[:,i]/len(df.columns)

# Hasil
np.round(distance, 3)
```

## No 2. Mencari data yang hampir sama untuk index 0-99

```python
# Definisi jarak terdekat data yang sama
near = np.zeros(100)

# Iterasi melalui jarak kolom matriks 
for i in range(100):
    
    # Nilai minimum index i, pilih 1 karena nilai 0 akan menjadi nilai 0 
    min_near = np.sort(distance[:,i])[1]
    near[i] = min_near
    
    # Mencari index yg memiliki nilai minimum untuk index i
    min_idx = np.where(distance[:,i] == min_near)
    print("Data indeks", i, "paling dekat dengan data", min_idx[0][0], "dengan kedekatan", distance[min_idx[0][0],i])

# Diperoleh data yang punya jarak terdekat 
min_near_idx = np.where(near == np.sort(near)[0])
print("Proximity paling dekat adalah data", min_near_idx)
```

## No 3. Menentukan jarak Minskowskian untuk $r = 1, 2, \infty$

```python
# Definisi titik untuk dimensi yang sama 
titik_1 = (-3, -2, -1, 0, 1, 2, 3)
titik_2 = (9, 4, 1, 0, 1, 4, 9)
n = len(titik_1)
```

```python
r=1
minkowski = 0
for i in range(n):
  minkowski += (abs(titik_1[i]-titik_2[i]))**r
minkowski = minkowski**(1/r)
minkowski
```
